```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
```

```python
image_name = "im.jpg"
image = imread(image_name)
plt.imshow(image)
```

```python
# rgb channel
r = image[:, :, 0]
g = image[:, :, 1]
b = image[:, :, 2]
```

```python
fig = plt.figure(figsize=(300, 300))

ax_r = fig.add_subplot(1, 3, 1)
plt.imshow(r)

ax_g = fig.add_subplot(1, 3, 2)
plt.imshow(g)

ax_b = fig.add_subplot(1, 3, 3)
plt.imshow(b)

plt.show()
```
